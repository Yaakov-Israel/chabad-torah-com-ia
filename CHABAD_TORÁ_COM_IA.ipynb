```python
import google.generativeai as genai

# Usaremos o mesmo modelo que tem funcionado bem.
# Lembre-se que 'gemini-1.5-flash-latest' é o nome do modelo.
modelo_detalhista = genai.GenerativeModel('gemini-1.5-flash-latest')

def resposta_em_niveis(pergunta_do_usuario):
    resposta_curta = None
    resposta_detalhada = None

    # Pedido para a resposta curta
    prompt_curto = f"Responda de forma muito sucinta e direta à seguinte pergunta: \"{pergunta_do_usuario}\""
    try:
        resposta_curta_ia = modelo_detalhista.generate_content(prompt_curto)
        resposta_curta = resposta_curta_ia.text
    except Exception as e:
        resposta_curta = f"Erro ao gerar resposta curta: {e}"

    # Pedido para a resposta detalhada (baseada na pergunta original)
    if resposta_curta and not "Erro" in resposta_curta: # Tenta detalhar se a curta funcionou
        prompt_detalhado = f"Considerando a pergunta \"{pergunta_do_usuario}\", e que uma resposta inicial curta seria algo como \"{resposta_curta}\". Agora, por favor, elabore sobre essa pergunta com mais detalhes, profundidade e, se possível, mencione aspectos relevantes ou fontes."
        try:
            resposta_detalhada_ia = modelo_detalhista.generate_content(prompt_detalhado)
            resposta_detalhada = resposta_detalhada_ia.text
        except Exception as e:
            resposta_detalhada = f"Erro ao gerar resposta detalhada: {e}"
    elif not resposta_curta:
         resposta_detalhada = "Não foi possível gerar a resposta detalhada pois a resposta curta falhou."


    return resposta_curta, resposta_detalhada

# Vamos testar!
pergunta1 = "Qual é o papel da Tzedaká (caridade) no Judaísmo?"

print(f"Pergunta: {pergunta1}\n")
curta1, detalhada1 = resposta_em_niveis(pergunta1)

print("--- Resposta Curta ---")
print(curta1)
print("\n--- Resposta Detalhada (se pedida) ---")
print(detalhada1)

print("\n-----------------------------------\n")

pergunta2 = "Explique o conceito de 'Tikun Olam'."
print(f"Pergunta: {pergunta2}\n")
curta2, detalhada2 = resposta_em_niveis(pergunta2)
print("--- Resposta Curta ---")
print(curta2)
print("\n--- Resposta Detalhada (se pedida) ---")
print(detalhada2)
```

```python
import google.generativeai as genai

# Usaremos o mesmo modelo que tem funcionado bem.
# 'gemini-1.5-flash-latest' é o nome do modelo.
modelo_chavruta_dialogo = genai.GenerativeModel('gemini-1.5-flash-latest')

def iniciar_dialogo_chavruta():
    print("--- INICIANDO CHAVRUTA INTERATIVA E CONTÍNUA ---")
    nome_sabio_chavruta = input("Com qual Sábio você gostaria de estudar hoje? (Ex: Rashi, Rambam, Hilel): ")
    tema_estudo_chavruta = input(f"Qual tema você gostaria de discutir com {nome_sabio_chavruta}? ")

    if not (nome_sabio_chavruta and tema_estudo_chavruta):
        print("Nome do Sábio ou tema não fornecidos. Encerrando Chavruta.")
        return

    print(f"\nExcelente! Iniciando seu estudo com {nome_sabio_chavruta} sobre '{tema_estudo_chavruta}'.")
    print("Para encerrar a Chavruta, digite 'sair', 'tchau' ou 'parar' a qualquer momento.")

    # Inicializa o chat com a instrução de sistema para a persona do Sábio
    # A API de chat do Gemini é mais adequada para conversas, gerenciando o histórico.
    # model.start_chat(history=[]) seria o ideal.
    # Como estamos fazendo uma simulação mais simples no Colab,
    # vamos construir o prompt com contexto a cada turno.

    # Instrução inicial para o modelo (contexto da Chavruta)
    instrucao_sistema = (
        f"Você é o {nome_sabio_chavruta}, um renomado mestre da Torá. "
        f"Eu sou seu parceiro de estudo (Chavruta). Estamos atualmente estudando e discutindo o tema: '{tema_estudo_chavruta}'. "
        f"Por favor, responda às minhas perguntas e comentários com a sua característica sabedoria, profundidade, "
        f"e estilo de pensamento, baseando-se nos ensinamentos judaicos relevantes. "
        f"Mantenha a persona do {nome_sabio_chavruta} durante toda a nossa conversa."
    )

    # Para simular o histórico de forma simples para o prompt do generate_content
    historico_para_prompt = [instrucao_sistema]

    print(f"\n{nome_sabio_chavruta}: Shalom! Sobre o que especificamente em '{tema_estudo_chavruta}' você gostaria de começar nossa discussão?")

    while True:
        interacao_usuario = input("\nVocê: ")

        if interacao_usuario.lower() in ["sair", "tchau", "parar", "chega"]:
            print(f"\n{nome_sabio_chavruta}: Que nosso aprendizado seja para elevação. Até nosso próximo estudo! Shalom.")
            break

        # Adiciona a fala do usuário ao histórico para o prompt
        historico_para_prompt.append(f"Meu parceiro de estudo (eu): {interacao_usuario}")

        # Monta o prompt completo
        prompt_completo = "\n".join(historico_para_prompt) + f"\nSua resposta como {nome_sabio_chavruta}:"

        try:
            resposta_modelo = modelo_chavruta_dialogo.generate_content(prompt_completo)
            texto_resposta_modelo = resposta_modelo.text

            print(f"\n{nome_sabio_chavruta}: {texto_resposta_modelo}")

            # Adiciona a resposta do modelo ao histórico para o prompt
            historico_para_prompt.append(f"Sua resposta como {nome_sabio_chavruta}: {texto_resposta_modelo}")

        except Exception as e:
            print(f"Ocorreu um erro na nossa Chavruta: {e}")
            print("Você pode tentar reformular ou, se o problema persistir, digite 'sair' para encerrar.")
            # Remove a última fala do usuário do histórico se a geração falhou para evitar repetição do erro
            if historico_para_prompt and "Meu parceiro de estudo (eu):" in historico_para_prompt[-1]:
                 historico_para_prompt.pop()


# Para iniciar a Chavruta, descomente a linha abaixo e execute a célula:
# iniciar_dialogo_chavruta()
```

```python
import google.generativeai as genai

# Usaremos o mesmo modelo que tem funcionado bem.
# 'gemini-1.5-flash-latest' é o nome do modelo.
modelo_chavruta_com_estilo = genai.GenerativeModel('gemini-1.5-flash-latest')

def iniciar_chavruta_linguagem_escolhida():
    print("--- INICIANDO CHAVRUTA COM ESCOLHA DE ESTILO DE LINGUAGEM ---")
    nome_sabio_escolhido = input("Com qual Sábio você gostaria de estudar hoje? ")
    tema_para_estudo = input(f"Qual tema você gostaria de discutir com {nome_sabio_escolhido}? ")

    estilo_pref = ""
    while estilo_pref not in ["tradicional", "moderna"]:
        estilo_pref = input(f"Você prefere que {nome_sabio_escolhido} use uma linguagem 'tradicional' ou 'moderna'? Digite sua escolha: ").lower()

    if not (nome_sabio_escolhido and tema_para_estudo):
        print("Faltou o nome do Sábio ou o tema. Encerrando a Chavruta.")
        return

    print(f"\nIniciando seu estudo com {nome_sabio_escolhido} sobre '{tema_para_estudo}', com linguagem {estilo_pref}.")
    print("Lembre-se: para encerrar, digite 'sair', 'tchau' ou 'parar'.")

    instrucao_estilo_linguagem = ""
    if estilo_pref == "tradicional":
        instrucao_estilo_linguagem = "Por favor, use uma linguagem que reflita a erudição e o estilo da sua época, mantendo a profundidade e, se apropriado, citando fontes ou usando termos específicos, mas de forma que um estudante dedicado possa compreender."
    else: # moderna
        instrucao_estilo_linguagem = "Por favor, adapte sua linguagem para ser mais moderna, clara e acessível a um estudante dos dias de hoje. Sinta-se à vontade para usar analogias ou explicações contemporâneas que ajudem no entendimento, sem perder a essência e a autenticidade dos seus ensinamentos."

    instrucao_base_para_ia = (
        f"Você é o Sábio {nome_sabio_escolhido}, um respeitado mestre da Torá. "
        f"Eu sou seu parceiro de estudo (Chavruta), e estamos explorando o tema: '{tema_para_estudo}'. "
        f"Mantenha sua persona e responda às minhas perguntas e comentários. {instrucao_estilo_linguagem}"
    )

    historico_para_geracao = [instrucao_base_para_ia]

    print(f"\n{nome_sabio_escolhido}: Shalom! Estou à disposição para começarmos nosso estudo sobre '{tema_para_estudo}', utilizando o estilo de linguagem que você selecionou. Qual é o seu primeiro pensamento ou questão?")

    while True:
        pergunta_do_usuario = input("\nVocê: ")
        if pergunta_do_usuario.lower() in ["sair", "tchau", "parar", "chega"]:
            print(f"\n{nome_sabio_escolhido}: Que este estudo nos aproxime da verdade. Até nosso próximo encontro! Shalom.")
            break

        historico_para_geracao.append(f"Meu parceiro de estudo (eu): {pergunta_do_usuario}")
        prompt_atual = "\n".join(historico_para_geracao) + f"\nSua resposta como {nome_sabio_escolhido}:"

        try:
            resposta_ia = modelo_chavruta_com_estilo.generate_content(prompt_atual)
            texto_resposta_ia = resposta_ia.text
            print(f"\n{nome_sabio_escolhido}: {texto_resposta_ia}")
            historico_para_geracao.append(f"Sua resposta como {nome_sabio_escolhido}: {texto_resposta_ia}")
        except Exception as e:
            print(f"Ocorreu um erro em nossa Chavruta: {e}")
            print("Você pode tentar de novo ou digitar 'sair' para encerrar.")
            if historico_para_geracao and "Meu parceiro de estudo (eu):" in historico_para_geracao[-1]:
                 historico_para_geracao.pop()

# Para iniciar a Chavruta com escolha de estilo, descomente a linha abaixo e execute a célula:
# iniciar_chavruta_linguagem_escolhida()
```

```python
import google.generativeai as genai

# Usaremos o mesmo modelo que tem funcionado bem.
# 'gemini-1.5-flash-latest' é o nome do modelo.
modelo_info_festas = genai.GenerativeModel('gemini-1.5-flash-latest')

def buscar_informacoes_festa(nome_da_festa):
    info_concisa_festa = None
    info_detalhada_festa = None

    # Gerando a informação concisa
    prompt_conciso_festa = f"Por favor, forneça uma descrição concisa da festa judaica de '{nome_da_festa}'. Inclua seu principal significado e, de forma geral, quando ela é celebrada no calendário judaico."
    try:
        resposta_concisa_modelo = modelo_info_festas.generate_content(prompt_conciso_festa)
        info_concisa_festa = resposta_concisa_modelo.text
    except Exception as e:
        info_concisa_festa = f"Desculpe, não consegui encontrar informações concisas sobre '{nome_da_festa}': {e}"

    # Gerando a informação detalhada, se a concisa funcionou
    if info_concisa_festa and not "Desculpe, não consegui encontrar" in info_concisa_festa:
        prompt_detalhado_festa = f"""Considerando a festa judaica de '{nome_da_festa}', por favor, elabore com mais detalhes.
Descreva as principais mitzvot (mandamentos) e costumes associados a ela.
Quais são as proibições importantes (halachot sobre trabalho, alimentação específica como chametz em Pessach, etc., se aplicável)?
Qual é o significado espiritual ou histórico mais profundo por trás das observâncias desta festa?"""
        try:
            resposta_detalhada_modelo = modelo_info_festas.generate_content(prompt_detalhado_festa)
            info_detalhada_festa = resposta_detalhada_modelo.text
        except Exception as e:
            info_detalhada_festa = f"Desculpe, não consegui encontrar informações detalhadas sobre '{nome_da_festa}': {e}"
    elif not info_concisa_festa:
        info_detalhada_festa = "Não foi possível buscar detalhes pois a informação inicial sobre a festa falhou."


    return info_concisa_festa, info_detalhada_festa

# --- Interação com o Usuário ---
print("🕎 --- Informações sobre as Festas Judaicas --- 🕎")
festa_desejada = input("Sobre qual festa judaica você gostaria de aprender? (Ex: Pessach, Shavuot, Rosh Hashanah, Yom Kipur, Sucot, Chanucá, Purim): ")

if festa_desejada:
    print(f"\nBuscando dados sobre {festa_desejada}...")
    print("Aguarde um momento, por favor.\n")

    conciso, detalhado = buscar_informacoes_festa(festa_desejada)

    print(f"--- Resumo Conciso de {festa_desejada} ---")
    print(conciso)

    print(f"\n--- Detalhes e Aprofundamento sobre {festa_desejada} ---")
    print(detalhado)
else:
    print("Nenhuma festa foi inserida. Tente novamente.")
```

```python
import google.generativeai as genai

modelo_super_aprofundamento = genai.GenerativeModel('gemini-1.5-flash-latest')

def gerar_resposta_inicial_v2(pergunta_usuario_original):
    prompt_curto = f"Forneça uma resposta inicial concisa e direta para: \"{pergunta_usuario_original}\""
    try:
        resposta = modelo_super_aprofundamento.generate_content(prompt_curto)
        return resposta.text
    except Exception as e:
        return f"Erro na resposta inicial: {e}"

def gerar_resposta_detalhada_v2(pergunta_original, instrucao_detalhe, resposta_curta_opcional=None):
    if instrucao_detalhe == "aprofundar_resposta_anterior" and resposta_curta_opcional:
        prompt_detalhado = f"""Sobre a pergunta \"{pergunta_original}\", e considerando a resposta inicial \"{resposta_curta_opcional}\".
Por favor, aprofunde essa linha de pensamento com mais detalhes, exemplos ou fontes."""
    elif instrucao_detalhe == "detalhar_pergunta_original":
        prompt_detalhado = f"""Sobre a pergunta original: \"{pergunta_original}\".
Por favor, forneça uma resposta completa e detalhada desde o início, explorando diferentes facetas,
mencionando fontes ou conceitos relevantes e oferecendo uma compreensão abrangente."""
    else:
        return "Instrução de detalhamento não reconhecida."

    try:
        resposta = modelo_super_aprofundamento.generate_content(prompt_detalhado)
        return resposta.text
    except Exception as e:
        return f"Erro ao gerar resposta detalhada: {e}"

print("💡 --- Pergunte ao Torá com IA (Aprofundamento Aprimorado) --- 💡")
minha_pergunta_v2 = input("Qual é a sua pergunta sobre Judaísmo? ")

if minha_pergunta_v2:
    print("\nBuscando uma resposta inicial...")
    resposta_inicial_v2 = gerar_resposta_inicial_v2(minha_pergunta_v2)
    print("\n--- Resposta Inicial ---")
    print(resposta_inicial_v2)

    if resposta_inicial_v2 and not "Erro na resposta inicial" in resposta_inicial_v2:
        escolha_usuario = ""
        while escolha_usuario not in ['s', 'r', 'n']:
            escolha_usuario = input("\nO que fazer agora? (s = aprofundar esta resposta, r = tentar resposta mais detalhada para pergunta original, n = nada): ").strip().lower()

        if escolha_usuario == 's':
            print("\nOk, aprofundando a resposta dada...")
            resposta_final_v2 = gerar_resposta_detalhada_v2(minha_pergunta_v2, "aprofundar_resposta_anterior", resposta_inicial_v2)
            print("\n--- Aprofundamento da Resposta Anterior ---")
            print(resposta_final_v2)
        elif escolha_usuario == 'r':
            print("\nOk, buscando uma resposta mais detalhada para sua pergunta original...")
            resposta_final_v2 = gerar_resposta_detalhada_v2(minha_pergunta_v2, "detalhar_pergunta_original")
            print("\n--- Resposta Detalhada da Pergunta Original ---")
            print(resposta_final_v2)
        else: # 'n'
            print("\nEntendido. Sem mais detalhes desta vez.")
    else:
        print("\nNão foi possível obter uma resposta inicial.")
else:
    print("Nenhuma pergunta foi feita.")
```
